Data Columns Description
* <b>battery_capacity:</b>
    * The capacity of the smartphone's battery
    * Measured in milliampere-hours (mAh), indicating how long the device can operate on a single charge.
    * type: int64
* <b>rear_primary_mp:</b>
    * The megapixel value of the primary rear camera.
    * Higher megapixels generally indicate better image resolution
    * type: float64
* <b>front_primary_mp:</b>
    * The megapixel value of the primary front (selfie) camera.
    * Important for selfies and video calls
    * type: int64
* <b>display_size_cm:</b>
    * The diagonal size of the device's display.
    * Measured in centimeters (cm), it indicates how large the screen
    * type : float64
* <b>display_quality1:</b>
    * The quality or resolution of the device's display.
    * Examples include HD, HD+, and Full HD+
    * type: object
* <b>ram:</b>
    * The amount of Random Access Memory (RAM) available in the device.
    * Measured in gigabytes (GB)
    * type: float64
* <b>rom:</b>
    * The internal storage capacity of the device.
    * Measured in gigabytes (GB)
    * type: float64
* <b>device_name:</b>
    * The specific model name of the smartphone.
    * Often includes the series or variant of the device
    * type: object
* <b>device_company:</b>
    * The brand or company that manufactures the device.
    * Examples include Realme, Samsung, Xiaomi
    * type: object
* <b>processor_brand:</b>
    * The brand of the device's processor.
    * Common brands include Qualcomm, MediaTek, and Exynos
    * type: object
* <b>processor_model:</b>
    * The specific model of the processor.
    * Defines the processing power and capabilities (e.g., Snapdragon 888)
    * type: object
* <b>review_count:</b>
    * The total number of user reviews received for the device.
    * Helps indicate popularity and user engagement
    * type :float64
* <b>warranty_years:</b>
    * The duration of warranty coverage provided by the manufacturer.
    * Typically measured in years
    * type :float64
* <b>rating:</b>
    * The average user rating for the device.
    * Usually represented on a scale (like 1 to 5 stars)
    * type :float64
* <b>price:</b>
    * The cost of the device.
    * Usually indicated in a specific currency 
    * type :int64
